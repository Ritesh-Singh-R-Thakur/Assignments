# K-Means clustering

```python
from sklearn.cluster import KMeans 
```

```python
km=KMeans()
```

```python
yp_kmeans=km.fit_predict(x)
yp_kmeans
```

```python
km.cluster_centers_
```

```python
#visulalize
plt.scatter(x[:,0],x[:,1],c=yp_kmeans,cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='black',s=100)
```

# Elbow Graph

```python
wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x=4,linestyle='--',color='red')
plt.xlabel('K-values')
plt.ylabel('wcss')
plt.title('Elbow Graph')
```

```python
#final Kmeans model
km=KMeans(n_clusters=4)
ypred_kmeans=km.fit_predict(x)
ypred_kmeans
```

```python
plt.scatter(x[:,0],x[:,1],c=ypred_kmeans,cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='black',s=150 )
```

```python
#silhouette score for kmeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,ypred_kmeans)
```

```python
df2=pd.read_csv('crime_data.csv')
df2.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
#Analyzing cluster
df2['cluster_kmeans']=ypred_kmeans
df2
```

```python
df2.groupby('cluster_kmeans').agg('mean')
```

# DBSCAN Clustering

```python
df3=pd.read_csv('crime_data.csv')
df3.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df3.boxplot()
```

```python
sns.boxplot(df3['Rape'])
```

```python
sns.pairplot(df3)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.5,min_samples=3)
```

```python
y_dbscan=db.fit_predict(x)
y_dbscan#DBSCAN shows all are outliers
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh=NearestNeighbors(n_neighbors=3).fit(x)
```

```python
d,index=neigh.kneighbors(x)
d
```

```python
distance=np.sort(d[:,1])
distance
```

```python
plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('Epsilon')
plt.title('K-Distance Graph')
```

```python
db1 = DBSCAN(eps=1.25,min_samples=3)
yp_dbscan=db1.fit_predict(x)
yp_dbscan
```

```python
plt.scatter(x[:,1],x[:,2],c=yp_dbscan,cmap='viridis')
```

```python
silhouette_score(x,yp_dbscan)
```

```python
df3['Cluster_DBSCAN']=yp_dbscan
```
