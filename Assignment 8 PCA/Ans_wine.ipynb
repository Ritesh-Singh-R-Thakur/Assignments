```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=250
sns.set_theme(style='darkgrid',palette="turbo_r")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('wine.csv')
df
```

```python
y=df.iloc[:,0:1]
y
```

```python
x=df.iloc[:,1:]
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
lg = LogisticRegression()
lg.fit(xtrain,ytrain)
ypred = lg.predict(xtest)

from sklearn.metrics import classification_report

print(classification_report(ytest,ypred))
```

```python
# Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x1=sc.fit_transform(x)
pd.DataFrame(x1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pcs = pca.fit_transform(x1)
pd.DataFrame(pcs)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=7)#using 90%  of data so selected 7 
pcs = pca.fit_transform(x1)
pcs
```

```python
data = pd.DataFrame(pcs, columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7'])
data['Type'] = y
data
```

```python
x5 = data.iloc[:,:-1]
y5 = data['Type']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x5,y5,test_size=0.2, random_state=1)
```

```python
lg = LogisticRegression()
lg.fit(xtrain,ytrain)
ypred = lg.predict(xtest)

print(classification_report(ytest,ypred))
```

```python
#observe we can see that after performing pca our over logistic regression model accurecy as increased from 94% to 97%
```

```python
from sklearn.cluster import KMeans
km=KMeans()
```

```python
# Elbow graph
```

```python
wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(x1)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x=3,linestyle='--',color='red')
plt.xlabel('K-values')
plt.ylabel('wcss')
plt.title('Elbow Graph')
```
