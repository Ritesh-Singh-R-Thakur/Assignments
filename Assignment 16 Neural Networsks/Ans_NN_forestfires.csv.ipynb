```python
#Importing libraries
# !pip install tensorflow
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the data
```

```python
df = pd.read_csv('forestfires.csv')
df
```

```python
df.drop(columns=['month','day'],inplace=True)
df
```

```python
df.describe()
```

```python
df.info()
```

# model building

```python
sc = StandardScaler()
x = df.drop(columns=['size_category'])
y = df['size_category']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y=le.fit_transform(y)
```

```python
y
```

```python
x = sc.fit_transform(x)
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20, random_state=1)
```

```python
#Initialize model
ann = Sequential()

#Add hidden layer
ann.add(Dense(units=10,activation='relu'))

#Add output layer
ann.add(Dense(units=1, activation='sigmoid'))

#Establish the connection between the layers
ann.compile(optimizer = 'adadelta',loss='binary_crossentropy',metrics=['accuracy'])

#Fit the data
ann.fit(xtrain,ytrain, epochs=100)
```

```python
ypred = ann.predict(xtest)
ypred
```

```python
ypred = ypred > 0.5
ypred
```

```python
print(classification_report(ytest,ypred))
```

```python
df['size_category'].value_counts()#imbalance data small and large value count should be same ratio accuracy would have been better.
                                   #lets try hyperparameter tuning.
```

```python
ann.evaluate(xtrain,ytrain)
```

# Hyperparameter tuning

```python
#!pip install -U keras-tuner
```

```python
import keras_tuner as kf
from keras_tuner import RandomSearch
```

```python
df.shape
```

```python
def hyper(hp):
    model = Sequential()
    counter = 0
    for i in range(hp.Int('num_layers', min_value = 1,max_value=10)):
        if counter ==0:
            model.add(Dense(hp.Int('units'+str(i),min_value=8,max_value=128,step=8),
                            activation = hp.Choice('activation'+str(i),values = ['relu','tanh','sigmoid']),
                            input_dim = 28))
    else:
        model.add(Dense(
              hp.Int('units'+str(i),min_value=8,max_value=128,step=8),
              activation = hp.Choice('activation'+str(i),values = ['relu','tanh','sigmoid'])))
    counter+=1
    model.add(Dense(units = 1,activation='sigmoid'))
    model.compile(optimizer = hp.Choice('optimizer',values = ['adam','rmsprop','sgd']),
                loss = 'binary_crossentropy',metrics = ['accuracy'])
    return model
```

```python
tuner = kf.RandomSearch(hyper,
                objective = 'val_accuracy',
                max_trials = 3,
                directory='new_project',  # change to a new directory
                project_name='tuner3')
```

```python
tuner.search(xtrain,ytrain,epochs = 5, validation_data = (xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
tuned_model = tuner.get_best_models(num_models=1)[0]
tuned_model
```

```python
tuned_model.summary()
```

```python
tuned_model.fit(xtrain,ytrain,epochs=50, validation_data=(xtest,ytest))
```

```python
tuned_model.evaluate(xtrain,ytrain)
```

```python
tuned_model.evaluate(xtest,ytest)
```

```python
def hyper(hp):
    model = Sequential()
    counter = 0
    for i in range(hp.Int('num_layers', min_value =3,max_value=12)):
        if counter ==0:
            model.add(Dense(hp.Int('units'+str(i),min_value=10,max_value=200,step=10),
                            activation = hp.Choice('activation'+str(i),values = ['relu','tanh','sigmoid']),
                            input_dim = 28))
    else:
        model.add(Dense(
              hp.Int('units'+str(i),min_value=10,max_value=200,step=10),
              activation = hp.Choice('activation'+str(i),values = ['relu','tanh','sigmoid'])))
    counter+=1
    model.add(Dense(units = 1,activation='sigmoid'))
    model.compile(optimizer = hp.Choice('optimizer',values = ['adam','rmsprop','sgd']),
                loss = 'binary_crossentropy',metrics = ['accuracy'])
    return model
```

```python
tuner = kf.RandomSearch(hyper,
                objective = 'val_accuracy',
                max_trials = 3,
                directory='new_project',  # change to a new directory
                project_name='tuner4')
```

```python
tuner.search(xtrain,ytrain,epochs = 10, validation_data = (xtest,ytest))
```
