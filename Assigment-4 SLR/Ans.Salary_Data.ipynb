# Salary_hike -> Build a prediction model for Salary_hike

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=150
%matplotlib inline
sns.set_theme(style='darkgrid',palette='turbo_r')
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Salary_Data.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
sns.scatterplot(x=df['YearsExperience'],y=df['Salary'])
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
df.head()
```

```python
model=smf.ols('Salary~YearsExperience',data=df).fit()
model
```

```python
model.params
```

```python
model.rsquared
```

```python
fig,index=plt.subplots(1,2)
sns.distplot(df['Salary'],ax=index[0])
sns.distplot(df['YearsExperience'],ax=index[1])
plt.tight_layout()
```

```python
df['Salary'].skew()
```

```python
df['YearsExperience'].skew()
```

# Log Transformation

```python
df['Log_of_Salary']=np.log(df['Salary'])
df['Log_of_YearsExperience']=np.log(df['YearsExperience'])
df
```

```python
fig,index=plt.subplots(2,2)
sns.distplot(df['Salary'],ax=index[0,0])
sns.distplot(df['Log_of_Salary'],ax=index[0,1])
sns.distplot(df['YearsExperience'],ax=index[1,0])
sns.distplot(df['Log_of_YearsExperience'],ax=index[1,1])
plt.tight_layout()
```

# sqrt Transforamtion

```python
df['sqrt_of_Salary']=np.sqrt(df['Salary'])
df['sqrt_of_YearsExperience']=np.sqrt(df['YearsExperience'])
df
```

```python
fig,index=plt.subplots(2,2)
sns.distplot(df['Salary'],ax=index[0,0])
sns.distplot(df['sqrt_of_Salary'],ax=index[0,1])
sns.distplot(df['YearsExperience'],ax=index[1,0])
sns.distplot(df['sqrt_of_YearsExperience'],ax=index[1,1])
plt.tight_layout()
```
